```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clima = pd.read_csv('../data/clima.csv', index_col=0)
ine_pernoctaciones=pd.read_csv('../data/ine_pernoctaciones.csv', index_col=0)
ine_viajeros=pd.read_csv('../data/ine_viajeros.csv', index_col=0)
ine_total=pd.read_csv('../data/ine_total.csv', index_col=0)
ign= pd.read_csv('../data/IGN.csv', index_col=0)
```

```python
# Visualizaci√≥n de la distribuci√≥n de la temperatura media
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(clima['Media_tmed'], kde=True, color='skyblue')
plt.title('Distribuci√≥n de Temperatura Media')

plt.subplot(1, 2, 2)
sns.boxplot(x='Periodo', y='Media_tmed', data=clima, palette='viridis')
plt.title('Variaci√≥n de Temperatura Media a lo largo del Tiempo')

plt.tight_layout()
plt.show();
```

```python
ine_pernoctaciones[['Viajeros_Pernoctaciones', 'Total']].describe().T
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Periodo', y='Viajeros_Pernoctaciones', hue='Provincia', data=ine_pernoctaciones, palette='viridis')
plt.title('Variaci√≥n de Pernoctaciones a lo largo del Tiempo y por Provincia')
plt.show();
```

```python
ine_viajeros[['Viajeros_Pernoctaciones', 'Total']].describe().T
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Periodo', y='Viajeros_Pernoctaciones', hue='Provincia', data=ine_viajeros, palette='viridis')
plt.title('Variaci√≥n de Viajeros a lo largo del Tiempo y por Provincia')
plt.show();
```

```python
ine_total[['Total']].describe().T
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Viajeros_Pernoctaciones', y='Total', data=ine_total, hue='Provincia', palette='viridis')
plt.title('Relaci√≥n entre Viajeros y Pernoctaciones por Provincia')
plt.show();
```

```python
# Visualizaci√≥n de la distribuci√≥n de la poblaci√≥n
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(ign['Poblaci√≥n'], kde=True, color='gold')
plt.title('Distribuci√≥n de la Poblaci√≥n')

# Diagrama de dispersi√≥n geoespacial
plt.subplot(1, 2, 2)
sns.scatterplot(x='Longitud', y='Latitud', size='Poblaci√≥n', data=ign, palette='viridis')
plt.title('Diagrama de Dispersi√≥n Geoespacial de la Poblaci√≥n')
plt.show();
```
